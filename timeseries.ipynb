```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import altair as alt
```

```python
questions = pd.read_csv("/Users/jessie/Desktop/UCSB/datafest_data/data/questions.csv")
```

```python
questions['AskedOnUtc'] = questions['AskedOnUtc'].str.split().str[0]
```

```python
questions['month'] = pd.to_datetime(questions['AskedOnUtc']).dt.month
```

```python
arizona = questions[questions['StateAbbr'] == 'AZ']
```

```python
arizona['month'].value_counts()
```

```python
arizona.to_csv("az.csv")
```

```python
arizona_ts = arizona.groupby(['month', 'Category']).size().unstack(fill_value=0).reset_index()
```

```python
arizona_ts['Consumer Financial Questions'] = arizona_ts['Consumer Financial Questions']/4.0
```

```python
arizona_ts['Education'] = arizona_ts['Education']/4.0
arizona_ts['Family and Children'] = arizona_ts['Family and Children']/4.0
arizona_ts['Health and Disability'] = arizona_ts['Health and Disability']/4.0
arizona_ts['Housing and Homelessness'] = arizona_ts['Housing and Homelessness']/4.0
arizona_ts['Income Maintenance'] = arizona_ts['Income Maintenance']/4.0
arizona_ts['Individual Rights'] = arizona_ts['Individual Rights']/4.0
arizona_ts['Juvenile'] = arizona_ts['Juvenile']/4.0
arizona_ts['Other'] = arizona_ts['Other']/4.0
arizona_ts['Work, Employment and Unemployment'] = arizona_ts['Work, Employment and Unemployment']/4.0
```

```python
arizona_ts
```

```python
alt.Chart(arizona_ts).mark_line().encode(
    x='month',
    y=alt.Y('Consumer Financial Questions', scale=alt.Scale(type='linear', domain=(0, 55)))
)

```

```python
alt.Chart(arizona_ts).mark_line().encode(
    x='month',
    y=alt.Y('Family and Children', scale=alt.Scale(type='linear', domain=(0, 55)))
)
```
