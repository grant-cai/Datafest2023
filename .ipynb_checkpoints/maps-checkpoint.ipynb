```python
import geopandas as gpd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
attorneys = pd.read_csv("/Users/jessie/Desktop/UCSB/datafest_data/data/attorneys.csv")
attorney_time_entries = pd.read_csv("/Users/jessie/Desktop/UCSB/datafest_data/data/attorneytimeentries.csv")
categories = pd.read_csv("/Users/jessie/Desktop/UCSB/datafest_data/data/categories.csv")
clients = pd.read_csv("/Users/jessie/Desktop/UCSB/datafest_data/data/clients.csv")
question_posts = pd.read_csv("/Users/jessie/Desktop/UCSB/datafest_data/data/questionposts.csv", on_bad_lines = "skip")
questions = pd.read_csv("/Users/jessie/Desktop/UCSB/datafest_data/data/questions.csv")
state_sites = pd.read_csv("/Users/jessie/Desktop/UCSB/datafest_data/data/statesites.csv")
subcategories = pd.read_csv("/Users/jessie/Desktop/UCSB/datafest_data/data/subcategories.csv")
```

```python
client_question = pd.merge(clients, questions, left_on='ClientUno', right_on='AskedByClientUno')
```

```python
client_question = client_question.drop(columns=['AskedByClientUno'])
```

```python
client_question.head()
```

```python
zip_map = gpd.read_file("/Users/jessie/Desktop/UCSB/datafest_data/zcta.zip")
```

```python
client_question = client_question.dropna()
```

```python
zip_map['ZCTA5CE10'] = zip_map['ZCTA5CE10'].astype(int)
client_question['PostalCode'] = client_question['PostalCode'].astype(int)
```

```python
merged_df_zip = zip_map.merge(client_question, left_on='ZCTA5CE10', right_on='PostalCode')
merged_df_zip= merged_df_zip.drop(columns=['PostalCode'])
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
merged_df_zip.plot(column='ZCTA5CE10', cmap='Reds', linewidth=0.1, edgecolor='gray', ax=ax, legend=True)
plt.title('Most Frequent Postal Codes in the US')
```

```python
states = gpd.read_file("/Users/jessie/Desktop/UCSB/datafest_data/s_22mr22.zip")
```

```python
states.head()
```

```python
merged_states = pd.merge(states, client_question, left_on='STATE', right_on='StateAbbr_x')
merged_states = merged_states.drop(columns=['StateAbbr_x'])
```

```python
state_counts = pd.DataFrame(merged_states['STATE'].value_counts()).reset_index()
```

```python
state_counts = state_counts.rename(columns={'STATE': 'count'})
```

```python
state_counts = state_counts.rename(columns={'index': 'STATE'})
```

```python
state_counts
```

```python
merged_df = states.merge(state_counts, left_on='STATE', right_on='STATE')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 8))

# Plot all states with their respective colors
merged_df.plot(column='count', cmap='OrRd', linewidth=0.5, ax=ax, edgecolor='0.5', legend=True)


# Hide the axis ticks and labels
ax.axis('off')


ax.set_xlim([-125, -65])  # adjust x limits
ax.set_ylim([25, 50])  # adjust y limits
plt.show()
```
